# Selenium 라이브러리
- 웹페이지를 자동으로 제어하는 라이브러리

```python
# 기본으로 내장된 라이브러리가 아니기 때문에 반드시 한번 설치!
!pip install selenium
```

```python
# webdriver = 웹을 통제하기 위한 라이브러리 ( 크롬 브라우저 역할 )
# Keys = 키보드 역할을 해주는 라이브러리
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
```

```python
## wb.Chrome() = 크롬창 실행 명령 / 보통 드라이버 변수에 담음
## .get("URL") = 원하는 페이지의 정보를 요청하는 기능

# 크롬드라이버 좀 실행시켜줘~
driver = wb.Chrome()
# 드라이버야 네이버 페이지 좀 가져와줘!
driver.get("http://www.naver.com")
```

```python
# 드라이버야 아이디가 ~ 인 요소 좀 찾아와!
search = driver.find_element_by_id("query")
```

```python
# 검색창에 코로나 값을 보내줘!
search.send_keys("코로나")
```

## 검색진행하기
1. 엔터를 입력해서 검색

```python
# 키보드로 엔터 사용
search.send_keys(Keys.ENTER)
```

```python
# 뒤로가라고 명령 ( 드라이버에게 )

driver.back()
```

```python
# 다시 새로운 페이지이므로 다시 서치를 찾고 값을 보내줘야함
search = driver.find_element_by_id("query")
search.send_keys("코로나")
```

2. 검색버튼을 클릭

```python
# 드라이버야 검색버튼 위치 찾아줘
btn = driver.find_element_by_id("search_btn")
```

```python
# 버튼을 클릭해줘!
btn.click()
```

```python
# 창 종료 ( 프로세서까지 닫음 )
driver.quit()
```

## 한솥도시락 사이트 메뉴명, 가격명 수집

```python
# 1. 브라우저 실행
# 2. 한솥도시락 URL로 이동
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
```

```python
# element 라고하면 한개만 가져옴
# -s를 하면 리스트로 가져옴
title = driver.find_elements_by_css_selector("h4.h.fz_03")
```

```python
for i in title :
    print(i.text)
```

```python
# 가격
price = driver.find_elements_by_css_selector("div.item-price > strong")
```

```python
for i in price :
    print(i.text)
```

```python
# 더보기 버튼을 클릭!
btn = driver.find_element_by_css_selector("a.c_05")
```

```python
btn.click()
```

# 예외처리 try, except
- try : 오류 나기전까지 실행할 로직
- except : 오류가 났을때 실행할 로직

```python
# time라이브러리
# 코드 중간중간에 쉬는시간 부여
import time
```

```python
try :
    for i in range(10) :
        btn.click()
        time.sleep(2)
except :
        print("더보기가 끝났습니다.")
```

```python
import pandas as pd
```

```python
# 모든 메뉴명과 가격명 수집
# 데이터프레임 제작
df_title = []

title = driver.find_elements_by_css_selector("h4.h.fz_03")
for i in title :
   df_title.append(i.text)
df_title
```

```python
df_price = []

price = driver.find_elements_by_css_selector("div.item-price > strong")
for i in price :
    df_price.append(i.text)
df_price
```

```python
infor = { '메뉴' : df_title , '가격' : df_price }
df= pd.DataFrame(infor)
df
```

```python
number = []
for i in range(1,112) :
    number.append(i)
infor['번호'] = number
infor
```

```python
df = pd.DataFrame(infor)
df.set_index('번호', inplace = True)
df
```

```python
# 데이터프레임 엑셀파일로 변환
# 인코딩
# 1. euc-kr
# 2. ""
# 3. utf-8-sig
df.to_csv("한솥.csv" , encoding = "euc-kr")
```

```python
# html 파일 저장
# 깨지면 인코딩 추가
df.to_html("한솥.html")
```
